```python
import torch
from parler_tts import ParlerTTSForConditionalGeneration
from transformers import AutoTokenizer
import soundfile as sf
import numpy as np
from scipy.io import wavfile
from pydub import AudioSegment
from sklearn.metrics import mean_squared_error
import torchaudio
import random
import os
import pandas as pd
from tqdm import tqdm
import nisqa
from nisqa.NISQA_model import nisqaModel
```

```python
device = "cuda:0" if torch.cuda.is_available() else "cpu"

model = ParlerTTSForConditionalGeneration.from_pretrained("parler-tts/parler-tts-mini-jenny-30H").to(device)
tokenizer = AutoTokenizer.from_pretrained("parler-tts/parler-tts-mini-jenny-30H")

prompt = "Hey, how are you doing today? My name is Jenny, and I'm here to help you with any questions you have."
description = "Jenny speaks at an average pace with an animated delivery in a very confined sounding environment with clear audio quality."

input_ids = tokenizer(description, return_tensors="pt").input_ids.to(device)
prompt_input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(device)

generation = model.generate(input_ids=input_ids, prompt_input_ids=prompt_input_ids)
audio_arr = generation.cpu().numpy().squeeze()
sf.write("parler_tts_out.wav", audio_arr, model.config.sampling_rate)
```

# Какие вообще метрики используются для голоса?


---



VoiceBox
*   **mean opinion scores (MOS)** (по ней нельзя сравнивать работы и даже разные исследования внутри одной работы)
*   измеряющая сходство на уровне сигнала - **mean cepstral distance (MCD)** - для синтеза речи и преобразования голоса и **signal-to-noise/distortion ratio (SNR/SDR)** (но эти метрики подразумевают детерминированность выходных данных относительно входых)
*  (ТОЧНОСТЬ) **word error rate (WER)** транскрипции синтезированной речи по отношению к входному тексту. Для сопоставимости используются общедоступные модели автоматического распознавания речи (ASR): для англоязычных HuBERT-L, предварительно обученный Librilight, Librispeech, для многоязычных - Whisper large-v2 (при зашумленном аудио такие модели могут просто не распознать слова, хотя реальной проблемы в них может не быть)
*   (СОГЛАСОВАННОСТЬ) эмбединги аудио контекста и синтезированной речи должны быть похожи (делают с помощью WavLM-TDCNN,
*   (РАЗНООБРАЗИЕ И КАЧЕСТВО) **Fréchet Speech Distance (FSD)** - сходство на уровне распределения в некотором пространстве признаков (через self-supervised wav2vec 2.0)
*   quality MOS и similarity MOS (пакет CrowdMOS Ribeiro et al)

Parler-TTS
*   гендерный классификатор  huggingface.co/alefiury/wav2vec2-large-xlsr-53-gender-recognition-librispeech
*   смотрят на “excellent recording quality” с помощью модели Torchaudio Speech Quality and Intelligibility Measures , которая оценивает **Wideband Perceptual Estimation of Speech Quality (PESQ), Short-Time Objective Intelligibility (STOI), and Scale-Invariant Signal-to-Distortion Ratio (SD-SRD)**

MELLE
*   аудио для теста из LibriSpeech test-clean set, отбор аудио от 4 до 10 секунд
*   WER -  Conformer-Transducer model3 and HuBERT-Large ASR model4
*  ** SIM (speaker similarity) ** (SIM-r измеряет сходство между синтезированной речью и речевой подсказкой, а SIM-o - сходство по отношению к исходной речевой подсказке, рекомендуют брать SIM-o так как SIM-r может быть не сопоставим между системами) - cosine distance между эмбедингами reference speech prompt и сгенерированной речи
*   **Inference time** (измерили время вывода для генерации 10-секундных сегментов речи, усредненное по десяти испытаниям)


NaturalSpeech3
*   speaker-similarity **SIM-O and SIM-R** (WavLM-TDCNN4)
* для качества речи (**UTMOS**)
* robustness - **WER** (ASR model - CTC-based HuBERT pre-trained on Librilight and fine-tuned on the 960 hours training set of LibriSpeech)

## **декомпозируйте понятие "оценка синтеза речи"**

Оценка синтеза речи должна проводиться по таким аспектам:

*   соответствие аудио и исходного текста
*   разборчивость произношения
*   наличие шума
*   соответсвие интонации
*   естественность (по сравнению с человеческой речью)

То есть оценка синтеза речи включает в себя определение правильности слов в аудио, правильности произношения, интонации, чистоты речи, похожести речи на речь человека.

```python
libri_test_clean_dir = './LibriSpeech/LibriSpeech/test-clean'

def get_audio_duration(audio_path):
    try:
        waveform, sample_rate = torchaudio.load(audio_path)
        return waveform.shape[1] / sample_rate
    except Exception as e:
        print(f"Error loading {audio_path}: {e}")
        return 0

def get_transcription(dir_path):
    for file in os.listdir(dir_path):
        if file.endswith(".trans.txt"):
            trans_file_path = os.path.join(dir_path, file)
            with open(trans_file_path, 'r') as f:
                return f.readlines()
    return None

def get_audio_files_durations_and_texts(root_dir):
    audio_files = []
    for root, _, files in os.walk(root_dir):
        flac_files = [file for file in files if file.endswith(".flac")]
        if flac_files:
            duration = 0
            transcriptions = get_transcription(root)
            for flac_file in flac_files:
                full_path = os.path.join(root, flac_file)
                file_duration = get_audio_duration(full_path)
                duration += file_duration
                audio_file_name = flac_file.replace(".flac", "")
                transcription = next((line for line in transcriptions if line.startswith(audio_file_name)), None)

                if transcription:
                    transcription_text = transcription[len(audio_file_name):].strip()
                    audio_files.append((full_path, file_duration, transcription_text))
                    print(f"Found audio file: {full_path}, Duration: {file_duration:.2f}, Transcription: {transcription_text[:30]}...")  # Показываем только первые 30 символов
                else:
                    print(f"Transcription not found for {full_path}")
    return audio_files

def filter_by_duration(audio_files, min_duration, max_duration):
    return [file for file in audio_files if min_duration <= file[1] <= max_duration]

def safe_random_sample(files, n):
    if len(files) < n:
        print(f"Warning: Not enough files in category. Requested: {n}, Available: {len(files)}. Returning all available files.")
        return files
    return random.sample(files, n)

def select_audio_files(audio_files, n_short=3, n_medium=3, n_long=3):
    short_files = filter_by_duration(audio_files, 0, 5)
    print(f"Number of short files: {len(short_files)}")
    selected_short = safe_random_sample(short_files, n_short)

    medium_files = filter_by_duration(audio_files, 5, 15)
    print(f"Number of medium files: {len(medium_files)}")
    selected_medium = safe_random_sample(medium_files, n_medium)

    long_files = filter_by_duration(audio_files, 15, float('inf'))
    print(f"Number of long files: {len(long_files)}")
    selected_long = safe_random_sample(long_files, n_long)

    selected_files = selected_short + selected_medium + selected_long
    return selected_files

audio_files = get_audio_files_durations_and_texts(libri_test_clean_dir)

selected_audio_files = select_audio_files(audio_files)

for audio_path, duration, transcription in selected_audio_files:
    print(f"Selected: {audio_path}, Duration: {duration:.2f} seconds, Transcription: {transcription}")
```

```python
model = ParlerTTSForConditionalGeneration.from_pretrained("parler-tts/parler-tts-mini-jenny-30H").to(device)
tokenizer = AutoTokenizer.from_pretrained("parler-tts/parler-tts-mini-jenny-30H")

generated_audio_files = []
original_transcriptions = []
original_audio = []

model_weights_path = "/content/NISQA/weights/nisqa_mos_only.tar"
output_dir = "/content/results"
os.makedirs(output_dir, exist_ok=True)

def generate_audio(prompt, description, index):
    input_ids = tokenizer(description, return_tensors="pt").input_ids.to(device)
    prompt_input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(device)

    generation = model.generate(input_ids=input_ids, prompt_input_ids=prompt_input_ids)

    audio_arr = generation.cpu().numpy()
    print(f"Generated audio shape: {audio_arr.shape}")

    if audio_arr.ndim == 1:
        audio_arr = audio_arr[np.newaxis, :]

    audio_arr = audio_arr.astype(np.float32)

    output_audio_path = f"/content/generated_audio_{index}.wav"
    sf.write(output_audio_path, audio_arr.T, model.config.sampling_rate)
    return output_audio_path

def evaluate_audio(wav_file_path):
    os.system(f'python /content/NISQA/run_predict.py --mode predict_file --pretrained_model {model_weights_path} --deg {wav_file_path} --output_dir {output_dir}')


    results_path = os.path.join(output_dir, "NISQA_results.csv")
    mos = pd.read_csv(results_path)['mos_pred'].mean()
    return mos

total_mos = 0
num_evaluated = 0

for file_path, duration, transcription in tqdm(selected_audio_files, desc="Processing Audio Files"):
    for desc_index, description in enumerate(descriptions_for_audio):

        audio_file_path = generate_audio(transcription, description, f"{file_path.split('/')[-1].replace('.flac', '')}_{desc_index}")

        generated_audio_files.append(audio_file_path)
        original_transcriptions.append(transcription)
        original_audio.append(file_path)

        mos = evaluate_audio(audio_file_path)
        total_mos += mos
        num_evaluated += 1

average_mos = total_mos / num_evaluated if num_evaluated > 0 else 0
print(f"Average MOS: {average_mos:.2f}")

print(f"Number of generated audio files: {len(generated_audio_files)}")
print(f"Number of original transcriptions: {len(original_transcriptions)}")
print(f"Number of original audio files: {len(original_audio)}")
```
